```python
# Import pandas data analytics library
import pandas as pd
```

```python
# Read in source CSV file of pizza sales
data = pd.read_csv('PizzaSales.csv')
```

```python
# Convert CSV to DataFrame
df = pd.DataFrame(data)
```

```python
# Display first 15 orders
df.head(15)
```

```python
# Create separate DataFrame with units sold per pizza type, ordered from most to least popular
popular_pizzas = df['pizza_name'].value_counts()
```

```python
# Display top 10 most popular pizza types
popular_pizzas.head(10)
```

```python
# Convert DataFrame to Excel file and store on local filesystem
popular_pizzas.to_excel('PopularPizzas.xlsx')
```
